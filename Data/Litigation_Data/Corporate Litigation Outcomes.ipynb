## Cleaning and Pre-processing

Start by dropping any row without a ticker:

```python
print(dfs.shape)
dfs = dfs.dropna(axis = 0, subset = ['Ticker'])
print(dfs.shape)
```

Creating a column for case type:

```python
dfs = dfs.reset_index(drop = True)
cols = []
for i in dfs.index:
    desc = list(dfs.iloc[i, 20:][dfs.iloc[i, 20:] > 0].index) #column names where categorical var is 1
    cols.append(desc)
    
dfs['Case_Type'] = cols
```

```python
# can have more than one case type, though less than half do
pd.Series([len(item) for item in dfs['Case_Type']]).value_counts()
```

```python
dfs['Case_Type'][:5]
```

Changing to dates to datetime datatype:

```python
dfs.loc[:, 'Exposure Began'] = pd.to_datetime(dfs.loc[:, 'Exposure Began'], format = '%Y-%m-%d')
dfs.loc[:, 'Exposure Ended'] = pd.to_datetime(dfs.loc[:, 'Exposure Ended'], format = '%Y-%m-%d')
dfs.loc[:, 'Case Began'] = pd.to_datetime(dfs.loc[:, 'Case Began'], format = '%Y-%m-%d')
dfs.loc[:, 'Case Ended'] = pd.to_datetime(dfs.loc[:, 'Case Ended'], format = '%Y-%m-%d')
dfs['Year'] = dfs['Case Began'].dt.year
```

```python
dfs.to_csv('Corporate_Lit_Outcomes_Clean.csv', index = False)
dfs
```

## Aggregating by Year

Filtering to defendants:

```python
defend = ['Defendant (Lead)', 'Defendant', 'Nominal Defendant', 'Third Party Defendant', 'Defendant (Consolidated)', 
          'Defendant (Lead) (Consolidated)','Nominal Defendant (Lead)','Nominal Defendant (Consolidated)', 
          'Third Party Defendant (Lead)','Third Party Defendant (Consolidated)']
```

```python
print(dfs.shape)
dfs = dfs[dfs['Named'].isin(defend)]
print(dfs.shape)
```

```python
# new dataframe for all cases by year and company
dfall =  dfs.groupby(['Ticker', 'Year']).count()['Case_Type']
dfall = dfall.reset_index()
dfall = dfall[dfall['Ticker'] != '*']
dfall = dfall[dfall['Year'] >= 2000]
dfall.rename(columns = {'Case_Type': 'Total_Cases'}, inplace = True)
dfall
```

```python
# new dataframe for securites cases by year and company
dfsum = dfs.groupby(['Ticker', 'Year']).sum()['Securities/Commodities/Exchange (NOS 850)']
dfsum = dfsum.reset_index()
dfsum = dfsum[dfsum['Ticker'] != '*']
dfsum = dfsum[dfsum['Year'] >= 2000]
dfsum.rename(columns = {'Securities/Commodities/Exchange (NOS 850)': 'Securities'}, inplace = True)
dfsum
```

```python
# new sparse temp df - blank rows for all combos of year/company
tics = list(dfsum['Ticker'].unique())
years = list(dfsum['Year'].unique())
tics.sort()
tica = tics * len(years)
tica.sort()
yearsa = years * len(tics)
sparse = pd.DataFrame({'Ticker':tica, 'Year': yearsa})
sparse.sort_values(by = ['Ticker', 'Year'])

# merging 
df_clean = sparse.merge(dfsum, how = 'left', on = ['Ticker', 'Year'])
df_clean = df_clean.merge(dfall, how = 'left', on = ['Ticker', 'Year'])
df_clean = df_clean.merge(dfs[['Ticker', 'Year']], how = 'left', on = ['Ticker', 'Year'])
df_clean = df_clean.drop_duplicates()
df_clean = df_clean.fillna(0)
df_clean = df_clean.merge(dfs[[ 'Ticker', 'CIK Code']],how = 'left',  on = ['Ticker']) # adding cik code to link to 
df_clean = df_clean[~df_clean.duplicated(subset = ['Ticker', 'Year'])] # dropping duplicates
df_clean
```

```python
df_clean.to_csv('Corporate_Lit_Outcomes_AggByYear.csv', index = False)
```
