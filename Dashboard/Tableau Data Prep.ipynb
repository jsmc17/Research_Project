```python
import pandas as pd
```

```python
df = pd.read_csv('../Russell3000/Russell3000Regression.csv')
df
```

```python
dft = pd.read_csv('../Russell3000/Russell-text-attrs.csv')
dft
```

```python
dfv = df.merge(dft, how = 'left', left_on = ['fyear', 'cik'], right_on = ['year', 'cik'] )
dfv
```

```python
ggroup = pd.read_excel('Industry_Code_Lookups.xlsx', sheet_name = 'ggroup')
dfv = dfv.merge(ggroup, how = 'left', left_on = 'ggroup', right_on = 'Code')
dfv
```

```python
dfv = dfv.rename(columns = {'Name': 'Industry Group'})
dfv
```

```python
dfv.to_csv('Cleaned_4_Tableau.csv', index = False)
```
